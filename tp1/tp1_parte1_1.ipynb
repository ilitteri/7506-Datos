```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def path_for(dataset: str) -> str:
    return f'/content/drive/MyDrive/{dataset}.csv'


review = pd.read_csv(path_for("Dataset/review"))
```

```python
correlacion_entre_reviews_funny_cool_y_useful = review[["useful", "funny", "cool"]].corr()
correlacion_entre_reviews_funny_cool_y_useful

plt.figure(dpi=150)
sns.heatmap(correlacion_entre_reviews_funny_cool_y_useful, cmap="gray_r")
```
